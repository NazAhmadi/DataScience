### Data Pre-processing and Cleaning


```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
csv_file_path = '/content/drive/MyDrive/DataScienceFinalProject/cleaned_crime_data.csv'
CrimeData = pd.read_csv(csv_file_path)
CrimeData.head()
CrimeData.shape
```

```python
CostOfLiving = pd.read_csv('/content/drive/MyDrive/DataScienceFinalProject/cost_of_living_us.csv')
CostOfLiving.head()
CostOfLiving.shape
```

```python
USCities = pd.read_csv('/content/drive/MyDrive/DataScienceFinalProject/uscities.csv')
USCities.head()
USCities.shape
```

```python
CrimeData.head()
```

```python
CrimeData['State'] = CrimeData['State'].str.replace(r'\d+', '', regex=True).str.strip()
CrimeData.head()
```

```python
CrimeData['City'] = CrimeData['City'].str.upper()
CrimeData.head()
```

```python
CostOfLiving.head()
```

```python
CostOfLiving['county'].unique().size
```

```python
USCities.head()
```

```python
USCities['city'] = USCities['city'].str.upper()
USCities['state_name'] = USCities['state_name'].str.upper()
USCities['county_name'] = USCities['county_name'].str.upper()
USCities.head()
```

```python
CrimeUSCities = pd.merge(CrimeData, USCities, left_on=['City', 'State'], right_on=['city', 'state_name'], how='inner')
CrimeUSCities.head()
```

```python
CrimeUSCities.shape
```

```python
CrimeUSCities.columns.unique()
```

```python
common_counties = set(CrimeUSCities['county_name']).intersection(set(CostOfLiving['county']))
print(f"Number of common counties: {len(common_counties)}")
print(f"Common counties: {common_counties}")
```

```python
CrimeUSCities.to_csv('MergedData(Crime and Cities).csv')
```

```python
# prompt: sort county name in ascending order

CrimeUSCities.sort_values(by='county_name', ascending=True, inplace=True)
CrimeUSCities['county_name'].unique()
```

```python
CostOfLiving.sort_values(by='county', ascending=True, inplace=True)
CostOfLiving['county'].unique()
```

```python
# prompt: remove the word county from CostOfLiving['county']

CostOfLiving['county'] = CostOfLiving['county'].str.replace(' County', '')
CostOfLiving['county'] = CostOfLiving['county'].str.replace(' Parish', '')
CostOfLiving['county'] = CostOfLiving['county'].str.upper()
CostOfLiving['county'].unique()
```

```python
common_counties = set(CrimeUSCities['county_name']).intersection(set(CostOfLiving['county']))
print(f"Number of common counties: {len(common_counties)}")
print(f"Common counties: {common_counties}")
```

```python
FinalMergedData = pd.merge(CrimeUSCities, CostOfLiving, left_on='county_name', right_on='county', how='inner')
FinalMergedData.head()
```

```python
FinalMergedData.shape
```

```python
FinalMergedData.columns.unique()
```

```python
columns_to_drop = ['state_id', 'lat', 'lng', 'timezone', 'zips', 'id', 'case_id', 'state', 'county']
FinalMergedData.drop(columns=columns_to_drop, inplace=True)
FinalMergedData.head()
```

```python
FinalMergedData.columns.unique()
```

```python
# prompt: count zero values in each column for each column

zero_counts = FinalMergedData.eq(0).sum(axis=0)
for column, count in zero_counts.items():
  print(f"Column: {column}, Zero values: {count}")
```

```python
columns_to_drop = ['Arson2', 'military','childcare_cost']
FinalMergedData.drop(columns=columns_to_drop, inplace=True)
FinalMergedData.head()
```

```python
FinalMergedData.columns.unique()
```
